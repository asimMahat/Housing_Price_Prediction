## Data Cleaning 

```python
# Finding out the columns with missing values with the number of missing values 

missing_values = (train_df.isnull().sum().sort_values(ascending=False))
print(missing_values[missing_values > 0])
```

```python
#  When 15% of the data is missing, we should delete the corresponding variable.  source - ( found in internet)

# dealing with missing data

# Removing the column having null values greater than 1 , that is except column Electrical

train_df = train_df.drop((missing_values[missing_values > 1]).index,1)
# train_df = train_df.drop(train_df.loc[train_df['Electrical'].isnull()].index)
train_df = train_df.dropna()
#Checking if there is any missing data left 
train_df.isnull().sum().max() 
```

```python
train_df.columns
```

```python
# Month sold , year sold and id does not have significance to the sales price
drop_columns = ["MoSold", "YrSold", "Id"]
train_df = train_df.drop(drop_columns, axis=1)
train_df.head()
```

```python
train_df.columns
```

## Looking at the distribution of data 
- Skewness and Kurtosis

```python
#looking at the distribution of target variable (SalesPrice ) 
sns.distplot(train_df['SalePrice'])
```

```python
print("The skewness of given distribution is : ", train_df['SalePrice'].skew())
```

## Result : 

- The distribution of our target variable is positively skewed . 
- In a positively skewed distribution :

  Mean > Median > Mode

- We can say that more houses are sold in less than the average price .

```python
## Kurtosis measures the shape of the curve or distribution. 
## If the value of kurtosis is high then the data contains more outliers 

print("The Kurtosis of given distribution is  : ",  train_df['SalePrice'].kurt())
```

- High kurtosis in a data set is an indicator that data has heavy tails or outliers.

### Finding categorical columns

```python
# Categorical column
cat_col = train_df.select_dtypes(include=['object']).columns.tolist() # Object indicates a column has text
len(cat_col)
```

## Using LabelEncoder from Scikit learn for categorical columns

```python
from sklearn.preprocessing import LabelEncoder

cat_encoders = { }

for i in cat_col:
    encoder = LabelEncoder()
    train_df[i]= encoder.fit_transform(train_df[i])
    cat_encoders[i]= encoder    
```
